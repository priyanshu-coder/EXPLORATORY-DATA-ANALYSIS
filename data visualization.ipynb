```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.arange(1,10,0.1)
y=2*x+5

plt.plot(x,y,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('beginner')
plt.show()
```

```python
x=np.arange(1,10,0.1)
y1=2*x+5
y2=4*x+5

plt.subplot(1,2,1)
plt.plot(x,y1,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('beginner')


plt.subplot(1,2,2)
plt.plot(x,y1,color='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('beginner')
plt.show()
```

```python
dic={'a':1,'b':2,'c':3,'d':4}
obj=list(dic.keys())
val=list(dic.values())
```

```python
obj,val
```

```python
#bar graph

plt.bar(obj,val,color='g')
plt.title("bar graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#scatter plot

x=[1,2,3,4,5,6]
y1=[5,6,3,1,2,7]
y2=[3,2,5,6,1,4]

plt.figure(figsize=(5,5))
plt.scatter(x,y1,s=100)
plt.scatter(x,y2,marker='3',s=200)
plt.legend(['y1','y2'])
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#histogram

x=[1,1,2,3,4,5,8,2,5,9,5,2,2,5,4,1,2,3,5,6]
plt.hist(x)
plt.show()
```

```python
#boxplot

a=[1,2,3,4,5,6,7,8,9]
b=[1,2,3,4,5,6,5,4,2]
c=[5,6,7,8,9,4,7,8,5]

li=list([a,b,c])
plt.boxplot(li)
plt.show()
```

```python
#violinplot

a=[1,2,3,4,5,6,7,8,9]
b=[1,2,3,4,5,6,5,4,2]
c=[5,6,7,8,9,4,7,8,5]

li=list([a,b,c])
plt.grid()
plt.violinplot(li,showmedians=True,showmeans=True)
plt.show()
```

```python
#piechart

plt.pie(val,labels=obj,autopct='%0.1f%%',shadow=True,explode=(0,0,0,0.1))
plt.show()
```

```python
#donut plot
plt.pie(val,labels=obj,autopct='%0.1f%%',shadow=True,radius=2,colors=['grey','black','red','green'])
plt.pie([1],colors='w',radius=1)
plt.show()
```

```python
#area plot

ele=[10,20,30,40,50,60,70,80,90]
limit=[1,1,2,5,7,5,1,6,8]

plt.stackplot(ele,limit)
plt.show()
```
